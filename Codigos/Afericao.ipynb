# Análise dos dados da Função **"Compra"**
## Com 100 Usuários

```python
import pandas as pd

# Load the two CSV files to inspect their contents
file_github_summary = 'https://raw.githubusercontent.com/AndreReis34/Trabalho-Final-de-ADS/refs/heads/main/tabelas/summary%20compra_100.csv'

# Load the data into pandas dataframes
summary_data = pd.read_csv(file_github_summary)

# Display the first few rows of each file to understand their structure
summary_data.head()
```

```python
# Análise detalhada das métricas fornecidas

# Análise das principais estatísticas da tabela de resumo (summary_data)
summary_stats = summary_data.describe()

# Separando algumas colunas relevantes para análise mais detalhada:
# Latência (Average), throughput, erro percentual (Error %), desvio padrão (Std. Dev.)
average_response_time = summary_data['Average'].mean()
throughput = summary_data['Throughput'].mean()
error_rate = summary_data['Error %'].apply(lambda x: float(x.strip('%'))).mean()
std_deviation = summary_data['Std. Dev.'].mean()

# Valores mínimos e máximos de tempo de resposta
min_response_time = summary_data['Min'].min()
max_response_time = summary_data['Max'].max()

avg_bytes = summary_data['Avg. Bytes'].mean()

{
    'Resumo Estatístico': summary_stats,
    'Tempo Médio de Resposta': average_response_time,
    'Throughput Médio': throughput,
    'Taxa de Erro Média (%)': error_rate,
    'Desvio Padrão Médio': std_deviation,
    'Tempo Mínimo de Resposta': min_response_time,
    'Tempo Máximo de Resposta': max_response_time,
    'Tamanho Médio dos Bytes': avg_bytes
}
```
